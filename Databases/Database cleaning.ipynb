```python
import pandas as pd
from rdkit import Chem
from rdkit.Chem import PandasTools
import numpy as np
from molvs import standardize_smiles
```

```python
TDB=pd.read_excel('Tautomer_database_release_3a.xlsx', na_values='nul',usecols=['Solvent', 'SMILES_1',  'Quantitative_ratio_1', 'SMILES_2',  'Quantitative_ratio_2'])
TDB
```

```python
TDB=TDB.query("Solvent=='Water' or Solvent=='Water-d2'")
TDB=TDB.dropna(subset=['Quantitative_ratio_2'],how='any')
TDB=TDB.reset_index(drop=True)
TDB
```

```python
for i in range(len(TDB)):
    if type(TDB.loc[i, 'Quantitative_ratio_1'])==str or TDB.loc[i, 'Quantitative_ratio_1']==1 or pd.isnull(TDB.loc[i, 'Quantitative_ratio_1']):
        TDB=TDB.drop(i)

TDB
```

```python
TDB.loc[19,'Quantitative_ratio_2']=100-TDB.loc[19,'Quantitative_ratio_1']
TDB=TDB.reset_index(drop=True)
TDB
```

```python
TDB['KT']=TDB['Quantitative_ratio_2']/TDB['Quantitative_ratio_1']
TDB=TDB[['SMILES_1', 'SMILES_2', 'KT']]
TDB

```

```python
TB=pd.read_excel('Tautobase.xlsx', usecols=['SMILES_1', 'SMILES_2', 'logK', '% (Tautomer1)',  'Solvent','Meas'])
TB
```

```python
TB=TB.query("Solvent=='Water'")
TB=TB.query("Meas=='E'")
TB=TB.reset_index(drop=True)
TB
```

```python
for i in range(len(TB)):
    if pd.isnull(TB.loc[i, '% (Tautomer1)']) and pd.isnull(TB.loc[i, 'logK']):
        TB=TB.drop(i)
        
TB=TB.reset_index(drop=True)
TB
```

```python
for i in range(len(TB)):
    if TB.loc[i, '% (Tautomer1)']==0 or TB.loc[i, '% (Tautomer1)']==100:
        TB=TB.drop(i)

TB=TB.reset_index(drop=True)
TB
```

```python
for i in range(len(TB)):
    if type(TB.loc[i,'logK'])==str:
        TB=TB.drop(i)

TB=TB.reset_index(drop=True)
TB
```

```python
for i in range(len(TB)):
    if pd.isnull(TB.loc[i, '% (Tautomer1)']):
        TB.loc[i, 'KT']=10**TB.loc[i, 'logK']
    if pd.isnull(TB.loc[i, 'logK']):
        TB.loc[i, 'KT']=(100-TB.loc[i, '% (Tautomer1)'])/(TB.loc[i, '% (Tautomer1)'])

TB
```

```python
TB.sample(10)
```

```python
TB=TB[['SMILES_1', 'SMILES_2', 'KT']]
TB
```

```python
taut=pd.concat([TDB, TB], ignore_index=True)
taut.index.name = 'TautID'
taut
```

```python
taut['KT'] = taut['KT'].astype("float64")
taut['logKT']=np.log10(taut['KT'])
taut
```

```python
taut.sample(10)
```

```python
ma=0
mi=0
c=0
for i in range(len(taut)):
    if taut.loc[i, 'logKT']>0:
        ma=ma+1
    if taut.loc[i, 'logKT']<0:
        mi=mi+1
    if taut.loc[i, 'logKT']==0:
        c=c+1
```

```python
print('Number of equilibriums where the product has a greater proportion:', ma)
print('Number of equilibriums where the reactive has a greater proportion:', mi)
print('Number of equilibriums where the two molecules are in the same proportion:', c)
```

```python
for i in range(len(taut)):
    if taut.loc[i, 'logKT']>0 or taut.loc[i, 'logKT']==0:
        taut.loc[i,'T1']=taut.loc[i,'SMILES_1']
        taut.loc[i,'T2']=taut.loc[i,'SMILES_2']
        taut.loc[i,'logK']=taut.loc[i,'logKT']
    if taut.loc[i, 'logKT']<0:
        taut.loc[i,'T1']=taut.loc[i,'SMILES_2']
        taut.loc[i,'T2']=taut.loc[i,'SMILES_1']
        taut.loc[i,'KT'] = 1/taut.loc[i,'KT']
        taut.loc[i,'logK'] = np.log10(taut.loc[i,'KT'])


taut
```

```python
taut=taut[['SMILES_1','SMILES_2','T1','T2', 'KT', 'logKT', 'logK']]
taut.sample(10)
```

```python
taut=taut[['T1','T2', 'logK']]
tauts=taut
taut
```

```python
for i in range(len(tauts)):
    tauts.loc[i,'T1']=standardize_smiles(tauts.loc[i,'T1'])
    tauts.loc[i,'T2']=standardize_smiles(tauts.loc[i,'T2'])

tauts
```

```python
T1=tauts[['T1']]
T1.to_csv('T1.csv')
```

```python
T2=tauts[['T2']]
T2.to_csv('T2.csv')
```

```python
logK=tauts[['logK']]
logK.to_csv('logK.csv')
```
