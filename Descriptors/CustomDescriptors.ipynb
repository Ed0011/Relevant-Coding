```python
import pandas as pd
import rdkit
from rdkit import Chem
from rdkit.Chem import PandasTools
from conjugation import *
```

```python
T1=pd.read_csv('/Users/ed/Documents/CBIO3/Tautómeros/Water/Databases/T1.csv')
T2=pd.read_csv('/Users/ed/Documents/CBIO3/Tautómeros/Water/Databases/T2.csv')
```

```python
T1 = conjbonds(T1, 'T1')
T1 = carbonyl(T1, 'T1')
T1 = amide(T1, 'T1')
T1 = PiDPiA(T1, 'T1')
T1 = FSP3(T1, 'T1')
T1 = PFI(T1, 'T1')
T1 = ArMsp3(T1, 'T1')
T1 = HBondDonors(T1, 'T1')
T1.set_index('TautID')
T1.drop('TautID', axis = 1, inplace= True)
T1.index.name = 'TautID'
T1.sample(10)
```

```python
T2 = conjbonds(T2, 'T2')
T2 = carbonyl(T2, 'T2')
T2 = amide(T2, 'T2')
T2 = PiDPiA(T2, 'T2')
T2 = FSP3(T2, 'T2')
T2 = PFI(T2, 'T2')
T2 = ArMsp3(T2, 'T2')
T2= HBondDonors(T2, 'T2')
T2.drop('TautID', axis = 1, inplace= True)
T2.index.name = 'TautID'
T2.sample(10)
```

```python
T1.to_csv('T1_custom.csv')
T2.to_csv('T2_custom.csv')
```
